Making a new data set for restaurants using the khush.csv file

```python
import pandas as pd
```

```python
df = pd.read_csv('new_data.csv')

# Split the 'categories' column into a list of categories
df['categories'] = df['categories'].str.split(',')

# Create a new column for each category
df_categories_expanded = df['categories'].apply(pd.Series)

# Replace the original 'categories' column with the expanded columns
df = pd.concat([df, df_categories_expanded], axis=1).drop('categories', axis=1)

# Rename the new columns
new_column_names = [f'category_{i}' for i in range(df_categories_expanded.shape[1])]
df.columns = list(df.columns[:-len(new_column_names)]) + new_column_names

# Save the modified DataFrame to a new CSV file
df.to_csv('modifed_data.csv', index=False)

df.head(5)
```

One hot encoding the categories coloumn

```python
# List of unique categories
categories = [
    'Nightlife', 'Bars', 'American (New)', 'American (Traditional)', 'Sandwiches', 'Italian', 
    'Pizza', 'Coffee & Tea', 'Seafood', 'Chinese', 'Mexican', 'Burgers', 'Desserts', 'Bakeries', 
    'Japanese', 'Salad', 'Asian Fusion', 'Sushi Bars', 'Vegetarian', 'Fast Food', 'Steakhouses', 
    'Diners', 'Mediterranean', 'Ice Cream & Frozen Yogurt', 'Juice Bars & Smoothies', 'Vegan', 
    'Thai', 'Barbeque', 'Lounges', 'Ethnic Food', 'Chicken Wings', 'Gluten-Free', 'Middle Eastern', 
    'Comfort Food', 'Bagels', 'Indian', 'Noodles', 'Soup', 'Vietnamese', 'Korean', 'Tapas/Small Plates', 
    'French', 'Latin American', 'Food Stands', 'Greek', 'Tex-Mex', 'Donuts', 'Caribbean', 'Halal', 'Bubble Tea', 
    'Szechuan', 'Ramen', 'Food Trucks', 'Southern', 'Tapas Bars', 'Tacos', 'Dance Clubs', 'Irish', 'Spanish'
]

# One-hot encode the categories
for category in categories:
    df[category] = df[['category_0', 'category_1', 'category_2']].apply(lambda row: category in row.values, axis=1)

# Drop the original columns
df = df.drop(['category_0', 'category_1', 'category_2'], axis=1)

# Save the new DataFrame to a new CSV file
df.to_csv('data_encoded.csv', index=False)
```

Creating another dataset named user 

```python
df = pd.read_csv(khush.csv)

# Create dummy variables for categories
categories_df = df['categories'].str.get_dummies(sep=',')

# Concatenate dummy variables with the original dataframe
df = pd.concat([df, categories_df], axis=1)

# Group by user_id and aggregate statistics
aggregated_df = df.groupby('user_id').agg({
    'review_id': 'count',
    'stars': ['mean'],
    **{col: 'sum' for col in categories_df.columns}
}).reset_index()

# Rename columns
aggregated_df.columns = ['user_id', 'stars_count', 'rating_mean', *categories_df.columns]

# Save the result to a new CSV file
output_file_path = 'user.csv'
aggregated_df.to_csv(output_file_path, index=False)
```
